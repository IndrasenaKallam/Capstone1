```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
# Reading the Data file 
crimes = pd.read_pickle('cleaned_ChicagoCrimes.pkl')
crimes.head(5)
```

```python
crimes[crimes['Year']==2013].District.nunique()
```

```python
blacks_majority_districts = [15,11,10,21,2,7,9,3,6,4,5]
crimes[crimes.District.isin(blacks_majority_districts)]
```

# Association of arrests with Districts

```python
# comparision of Arrest and Non Arrest with District 1 and Rest all
blacks_majority_districts = [15,11,10,21,2,7,9,3,6,4,5]
table=pd.crosstab(crimes['Arrest'],crimes['District'].isin(blacks_majority_districts))
a = table[0][0]
c = table[1][0]
b = table[0][1]
d = table[1][1]
print(table)
print(a,b,c,d)
```

```python
# hypothesis Testing :
# H0 : There is no Association with arrests between the District 1 and rest 
# H1 : There is an association with arrest between the District 1 and rest
# chi-squared test with similar proportions
# contingency table
print("The Contingency Table")
print("False Represents the District's with less black's population")
print('*****************************TABLE****************************************')
print(table)
stat, p, dof, expected = chi2_contingency(table)
print('************************Degree of Freedom***************************************')
print('dof=%d' % dof)
print('************************Expected Values***************************************')
print(expected)
# interpret test-statistic
prob = 0.95
critical = chi2.ppf(prob, dof)
print('************************Results of chi2.ppf***************************************')
print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
print('************************Decision based on the  static value ***********************')
if abs(stat) >= critical:
	print('Dependent (reject H0)')
else:
	print('Independent (fail to reject H0)')
# interpret p-value
print('************************Decision based on the  p value ***********************')
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
```
